#### Kat Sayısı Sütunu İçin

```python
bk["Kat Sayısı"]=non_null["Kat Sayısı"]
```

```python
dolu=bk[bk["Kat Sayısı"].notnull()] #we seperate the df into two. If survived value is not null then it becomes our train set  
bos=bk[bk["Kat Sayısı"].isnull()]
dolu.head()
y=dolu["Kat Sayısı"] #dolu dataframes Survived value is our y_train
x=dolu.drop(["Kat Sayısı"],axis=1)
bos=pd.get_dummies(bos,drop_first=True) 
bos.shape
```

```python
models=[LR,R,L,E,GBR,ETR]
def model(models,x,y):
    row_list=[]
    for i in range(len(models)):
        model=models[i]
        #modelname=['gbc', 'd', 'k', 'g','b','r','log'] #names of models we will use them as indexes
        model.fit(x,y)
        tahmin=model.predict(x)
        
        #r2_score(y_test,y_pred_ridge)
        #creating a dictionary to calculate scores for each model in the models list
        data={'r2_score':(mean_squared_error(y,tahmin))**0.5,
              'mean_squared_error':r2_score(y,tahmin),
             }
        
        row_list.append(data) #appending the dictionary to a list to turn it into a dataframe
        
        #printing the scores
       # print(accuracy_score(y,tahmin),precision_score(y,tahmin),recall_score(y,tahmin),f1_score(y,tahmin))
        #print(confusion_matrix(y,tahmin))
       # print(classification_report(y,tahmin))
    
    #adding each element of row_list to the df dataframe (each element of row_list is a row of df )
    df = pd.DataFrame(row_list, index =['LR', 'R', 'L', 'E','GBR','ETR']) 
    return df
```
